VARIANT 12 - f(x) = 4cos(x) + 0.3x = 0

№1

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return np.cos(x)
def f2(x):
  return -0.3*x/4

x1 = np.linspace(-20,20, 10000)
y1 = f1(x1)
x2 = np.linspace(-20,20,10)
y2 = f2(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0, color='k', linestyle='-')
plt.grid(True)
plt.title('График f(x)=cos(x) и f(x)=-3x/40')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
from math import *

def f(x):
  return 4*cos(x) + 0.3*x

epselant = 1
a = pi/3
b = 2*pi/3
while epselant > 0.001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  epselant = (b - a)/2
globalAnswer = (a + b)/2
print(round(globalAnswer,3))
```

```python
from math import *

def f(x):
  return acos(-0.3*x/4)

x_n1 = 1
x_n2 = 1.65
count = 1
epselant = 10**(-6)
print(f'0 --- {x_n2}')
while abs(x_n2 - x_n1) > epselant:
    x_n1 = x_n2
    x_n2 = round(f(x_n1), 6)
    print(f'{count} --- {x_n2}')
    count += 1
print(f'ANSWER {x_n2}')
```

```python
from math import *
def f(x):
  y = 4*cos(x) + 0.3*x
  return y
xl = float(input('Введите левый х: '))
xr = float(input('Введите правый х: '))
eps = float(input('Введите эпселант: '))
d = (xl-xr)/2
n = 0
c = xl
print(f'n = {n}   xl = {xl}   xr = {xr}')
while abs(d) > eps:
  xl = round(xl - f(xl)/(3*xl**2 + 6*xl - 24), 6)
  xr = round((c*f(xr) - xr*f(c))/(f(xr) - f(c)), 6)
  y = (xl+xr)/2
  d = (xl-xr)/2
  n += 1
  print(f'n = {n}   xl = {xl}   xr = {xr}')
x = y
print(f'x = {x}  n = {n}')
```

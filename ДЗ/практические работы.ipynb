практическая работа №2

```python
from math import *
def f(x):
  answer = 2**x - 2*cos(x)
  return answer
epselant = 1
a = 0.5
b = 1
while epselant > 0.001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  epselant = (b - a)/2
globalAnswer = (a + b)/2
print(globalAnswer)
```

практическая работа №3

```python
from math import *
def f(x):
    answer = x + sin(x) - 0.2*x
    return answer
x_n1 = 0
x_n2 = 2.6
count = 1
epselant = 10**(-3)
print(f'0 --- {x_n2}')
while abs(x_n2 - x_n1) > epselant:
    x_n1 = x_n2
    x_n2 = round(f(x_n1), 3)
    print(f'{count} --- {x_n2}')
    count += 1
print(f'ANSWER {x_n2}')
```

## Упражнения параграфов учебника по ЧМ

2.2

сам код для 2.2

```python
from math import *

def f(x):
  return x*sin(x) - 1

epselant = 1
a = 1
b = 1.5
while epselant > 0.0001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  epselant = (b - a)/2
globalAnswer = (a + b)/2
print(round(globalAnswer,3))
```

```python
from math import *

def f(x):
  return x*sin(x) - 1

epselant = 1
a = 1
b = 1.5
while epselant > 0.0001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
    print(f'[a,c]  |   f(a)=f({a:.4f})= {f(a):.4f}   |   f(c)=f({c:.4f})= {f(c):.4f}')
  else:
    a = c
    print(f'[c,b]  |   f(c)=f({c:.4f})= {f(c):.4f}   |   f(b)=f({b:.4f})= {f(b):.4f}')
  epselant = (b - a)/2
globalAnswer = (a + b)/2
print()
print(f'ANSWER --- {round(globalAnswer,3)}')
```

```python
from math import *

def f(x):
  return 1/sin(x)

x_n1 = 1.5
x_n2 = 1.1
count = 1
epselant = 10**(-5)
print(f'0 --- {x_n2}')
while abs(x_n2 - x_n1) > epselant:
    x_n1 = x_n2
    x_n2 = round(f(x_n1), 5)
    print(f'{count} --- {x_n2}')
    count += 1
print(f'ANSWER {x_n2}')
```

```python
from math import *
def f(x):
  y = x**3 + x**2 - 24*x + 1
  # y = sin(2*x)-log(x)   # это  из примера
  return y
xl = float(input('Введите левый х: '))
xr = float(input('Введите правый х: '))
eps = float(input('Введите эпселант: '))
d = (xl-xr)/2
n = 0
c = xl
print(f'n = {n}   xl = {xl}   xr = {xr}')
while abs(d) > eps:
  xl = round(xl - f(xl)/(3*xl**2 + 6*xl - 24), 6)
  # xl = xl - f(xl)/(2*cos(2*xl)-1/xl)    # это  из примера
  xr = round((c*f(xr) - xr*f(c))/(f(xr) - f(c)), 6)
  y = (xl+xr)/2
  d = (xl-xr)/2
  n += 1
  print(f'n = {n}   xl = {xl}   xr = {xr}')
x = y
print(f'x = {x}  n = {n}')
```

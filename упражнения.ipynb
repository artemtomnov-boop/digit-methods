## Упражнения параграфов учебника по ЧМ

1.1

```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return np.sin(x)
def f2(x):
  return 1/x

x1 = np.linspace(-20,20, 10000)
y1 = f1(x1)
x2 = np.linspace(-20,20,100)
y2 = f2(x2)

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0, color='k', linestyle='-')
plt.grid(True)
plt.title('График f(x)=sin(x) и f(x)=1/x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
from math import *

def f(x):
  y = x*sin(x) - 1
  return y

eps = 1
a = 1
b = 1.2
while eps > 0.0001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  eps = (b - a)/2
answer = (a + b)/2
print(round(answer,3))
```

```python
from math import *

def f(x):
  y = 1/sin(x)
  return y

x1 = 1
# х2 = х0
x2 = 1.1
c = 1
eps = 10**(-5)
print(f'0 --- {x2}')
while abs(x2 - x1) > eps:
    x1 = x2
    x2 = round(f(x1), 5)
    print(f'{c} --- {x2}')
    c += 1
print(f'Ответ: {x2}')
```

```python
from math import *
def f(x):
  y = x**3 + x**2 - 24*x + 1
  return y
xl = float(input('Введите левый х: '))
xr = float(input('Введите правый х: '))
eps = float(input('Введите эпселант: '))
d = (xl-xr)/2
n = 0
c = xl
print(f'n = {n} | xl = {xl} | xr = {xr}')
while abs(d) > eps:
  xl = round(xl - f(xl)/(3*xl**2 + 6*xl - 24), 6)
  xr = round((c*f(xr) - xr*f(c))/(f(xr) - f(c)), 6)
  y = (xl+xr)/2
  d = (xl-xr)/2
  n += 1
  print(f'n = {n} | xl = {xl} | xr = {xr}')
x = y
print(f'x = {x}  n = {n}')

```
